## Bar Charts

- A bar chart is used when we want to compare metric values across different subgroups of the data. If we have a greater number of groups, a bar chart is preferred over a column chart.

- We want to visualise the total revenue or sales generated for each region


```python
plt.bar(df["Satisfaction"], df["Age"], color = ["beige", "wheat", "navajowhite", "tan", "darkkhaki"])
#plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(df["Satisfaction"], df["Age"], color = ["beige", "wheat", "navajowhite", "tan", "darkkhaki"])
plt.show()
```

```python
Age_count = df.groupby("Satisfaction")["Age"].count().to_frame().reset_index()
Age_count
```

```python
plt.bar(Age_count["Satisfaction"], Age_count["Age"], color = ["lightsteelblue", "lightsteelblue", "lightsteelblue", "slategrey", "lightsteelblue"])
```

```python
income = df.groupby("Satisfaction")["Income"].mean().to_frame().reset_index()
```

```python
income
```

```python
plt.bar(income["Satisfaction"], income["Income"], width = 0.90, color = ["lightsteelblue", "lightsteelblue", "slategrey", "lightsteelblue", "lightsteelblue"])
```

## Bubble Charts

```python
np.random.seed(42)
N = 100
x = np.random.normal(170, 20, N)
y = x + np.random.normal(5, 25, N)
colors = np.random.rand(N)
area = (25 * np.random.rand(N))**2
```

```python
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Colors': colors,
    "bubble_size":area})
df.head()
```

```python
plt.style.use('ggplot')
plt.scatter('X', 'Y', s='bubble_size',alpha=0.5, data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)
```

```python
plt.scatter('X', 'Y',s='bubble_size',c='Colors',alpha=0.6, data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)


#plt.scatter(df.X, df.Y, s=df.bubble_size, color=["red", "green", "blue", "grey", "green"])
```

```python
plt.style.use('ggplot')
plt.scatter('X', 'Y',alpha=0.5, c='Colors', data=df)
plt.xlabel("X", size=16)
plt.ylabel("y", size=16)
```

# Subplots
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html

```python
x = np.arange(1,50,2)
y = np.power(x,2)
z = np.power(x, 3)
w = np.sin(x)
```

```python
fig, ax = plt.subplots(3, figsize=(10, 5))

ax[0].plot(x, y, marker="o")
ax[0].set_ylabel("square")
ax[0].set_xticklabels([])


ax[1].plot(x, z, c="salmon")
ax[1].set_xticklabels([])


ax[2].plot(x, w, c= "lime")

#ax.set_title('A single plot')
```
